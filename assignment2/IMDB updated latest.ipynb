```python
url = 'https://www.imdb.com/chart/top?ref_=ft_250'
```

```python
import re
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get(url)
```

```python
html = response.text
```

```python
soup = BeautifulSoup(html)
```

```python
movies = soup.select('tbody tr')
```

```python
data = []
a=0
for movie in movies:
                poster = movie.select_one('.posterColumn')
                score = poster.select_one('span[name="ir"]')['data-value']
            
                movie_link = movie.select_one('.titleColumn').select_one('a')['href']
                year_str = movie.select_one('.titleColumn').select_one('span').get_text()
                year_pattern = re.compile('\d{4}')
                year = int(year_pattern.search(year_str).group())
                id_pattern = re.compile(r'(?<=tt)\d+(?=/?)')
                movie_id = id_pattern.search(movie_link).group()
                movie_name = movie.select_one('.titleColumn').select_one('a').string
                a=a+1
                
                data.append([a,movie_name,year,score,movie_id,movie_link])
```

```python
data
```

```python
import csv
with open ('top_movies.csv','w') as f:
    mywriter = csv.writer(f)
    header = ['ranking','movie','year','score','id','link']
    mywriter.writerow(header)
    mywriter.writerows(data)
```

```python
import csv
urls = []
with open('top_movie_urls.csv', 'r') as f:
    f_csv = csv.reader(f)
    for row in f_csv:
        row[1]="http://www.imdb.com/title/tt" + row[1]
        urls.append(row[1])
```

```python
urls
```

```python
data2=[]
```

```python
url=urls[14]
```

```python
def papapa(b):
    url=urls[b]
    r = requests.get(url)
    html2 = r.text
    soup2 = BeautifulSoup(html2,"html.parser")
    credit_summary = soup2.find('div',attrs={'class':'plot_summary'})  
    summary_text = credit_summary.find('div',attrs={'class':'summary_text'})
    credit = credit_summary.find_all('a')
    Director = credit[0].text
    Writer = credit[1].text
    keywords0 = soup2.find_all('span',attrs={'class':'itemprop'})
    keywords = [keywords0[0].text,keywords0[1].text,keywords0[2].text,keywords0[3].text,keywords0[4].text] 
    genre00 = soup2.find_all('div',attrs={'class':'see-more inline canwrap'})
    genre0=genre00[1].find_all('a')
    genre=[]
    for x in genre0:
      x = x.text
      genre.append(x)
    casts00 = soup2.find('table',attrs={'class':'cast_list'})
    cast0 = casts00.find_all('img')
    casts=[]
    for cast in cast0:
      cast = cast.get('title')
      casts.append(cast)
    a = soup2.find_all('div',attrs={'class':'txt-block'})
    Budget=a[9].text.replace('\n','')
    Budget=Budget[9:].replace('            (estimated)','')
    Gross_USA=a[11].text.replace('\n','')
    Gross_USA=Gross_USA[13:].replace(' ','end')
    y=Gross_USA.find(',end')
    Gross_USA=Gross_USA[0:y]
    Cumulative_W_G=a[12].text.replace('\n','')
    Cumulative_W_G=Cumulative_W_G[29:].replace(' ','end')
    z=Cumulative_W_G.find(',end')
    Cumulative_W_G=Cumulative_W_G[0:z]
    data2.append([Director,Writer,keywords,summary_text,genre,casts,Budget,Gross_USA,Cumulative_W_G])
    return  url,html2,soup2,credit_summary,summary_text,credit,Director,Writer,keywords0,keywords,genre00,genre0,genre0,genre,casts00,cast0,casts,a,Budget,Gross_USA,Cumulative_W_G,data2
```

```python
papapa(3)
```

```python
for b in (1,250): 
        data2.append(papapa(b))
```
