```python
url = 'https://www.imdb.com/chart/top'
```

```python
url
```

```python
import requests
```

```python
r = requests.get(url)
r.text
```

```python
from bs4 import BeautifulSoup
```

```python
mypage = BeautifulSoup(r.text)
```

```python
type(mypage)
```

```python
mymovie=mypage.find('tbody',attrs={'class':'lister-list'})
```

```python
items = mymovie.find_all('tr')
```

```python
data = []
for item in items:
    title = item.find('td',attrs={'class':'titleColumn'}).text.strip().replace('\n','')
    rating = item.find('td',attrs={'class':'ratingColumn imdbRating'}).text.strip()
    data.append([title,rating])
```

```python
data
```

```python
import csv
with open ('top_movie.csv','w') as f:
    mywriter = csv.writer(f)
    header = ['Movie','Rating']
    mywriter.writerow(header)
    mywriter.writerows(data)
```

```python
def get_one_list(ranktitle):
    mytd = ranktitle.find_all('td')
    data = []
    for mytd in mytd:
        data.append([mytd.text.strip(), mytd.attrs['class'],])
    return data
```
